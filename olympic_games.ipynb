```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset

athletes = pd.read_csv('G:\Projects\Olympic Dataset/athlete_events.csv')
regions = pd.read_csv('G:\Projects\Olympic Dataset/noc_regions.csv')
```

```python
# Join the dataframes

athletes_sat = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_sat.head()
```

```python
#Total no. of each athletes in every year.
feamale_participants = athletes_sat[(athletes_sat.Sex=='F') & (athletes_sat.Season=='Summer')] [['Sex', 'Year']]
feamale_participants = feamale_participants.groupby('Year').count().reset_index()
feamale_participants.head(8)
```

```python
feamale_olympics = athletes_sat[(athletes_sat.Sex=='F') & (athletes_sat.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x = 'Year', data= feamale_olympics, palette="Set1")
plt.title('Female participation')
```

```python
part = feamale_olympics.groupby('Year') ['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Atheletes versus time')
```

```python
male_olympics = athletes_sat[(athletes_sat.Sex=='M') & (athletes_sat.Season=='Summer')]
```

```python
part1 = male_olympics.groupby('Year') ['Sex'].value_counts()
part2 = feamale_olympics.groupby('Year') ['Sex'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'M'].plot(color = 'blue')
part2.loc[:,'F'].plot(color = 'pink')
plt.title('Plot of Male and Female Atheletes versus time')
```

```python
gold_Medals = athletes_sat[(athletes_sat.Team == 'United States')]
gold_Medals.head()
```

```python
#Gold, Silver and Bronze Medal athletes In USA

gold_Medals = athletes_sat[(athletes_sat.Team == 'United States')]
silver_Medals = athletes_sat[(athletes_sat.Team == 'United States')]
bronze_Medals = athletes_sat[(athletes_sat.Team == 'United States')]
```

```python
part1 = gold_Medals.groupby('Year') ['Medal'].value_counts()
part2 = silver_Medals.groupby('Year') ['Medal'].value_counts()
part3 = bronze_Medals.groupby('Year') ['Medal'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'Gold'].plot(color = 'gold')
part2.loc[:,'Silver'].plot(color = 'silver')
part3.loc[:,'Bronze'].plot(color = 'brown')
plt.title('Plot of Gold, Silver and Bronze Medal versus time in United States')
```

```python
#Gold, Silver and Bronze Medal athletes In France

gold_Medals = athletes_sat[(athletes_sat.Team == 'France')]
silver_Medals = athletes_sat[(athletes_sat.Team == 'France')]
bronze_Medals = athletes_sat[(athletes_sat.Team == 'France')]
```

```python
part1 = gold_Medals.groupby('Year') ['Medal'].value_counts()
part2 = silver_Medals.groupby('Year') ['Medal'].value_counts()
part3 = bronze_Medals.groupby('Year') ['Medal'].value_counts()
plt.figure(figsize=(20,10))
part1.loc[:,'Gold'].plot(color = 'gold')
part2.loc[:,'Silver'].plot(color = 'silver')
part3.loc[:,'Bronze'].plot(color = 'brown')
plt.title('Plot of Gold, Silver and Bronze Medal versus time in France')
```

```python
silvermedals = athletes_sat[(athletes_sat.Medal == 'Silver')]
silvermedals = silvermedals[np.isfinite(silvermedals['Age'])]
silvermedals.head()
```

```python
#Silver Medal under 15
silvermedals['ID'][silvermedals['Age'] <15].count()
```

```python
game_incident = silvermedals['Sport'][silvermedals['Age'] < 15]
game_incident
```

```python
# Plot for Sports_incidents
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(game_incident)
plt.title('Silver Medals for Athletes under 15 years')
```

```python
# Silver medals from each country
silvermedals.Region.value_counts().reset_index(name='Medal').head(12)
```

```python
totalSilverMedals = silvermedals.Region.value_counts().reset_index(name='Medal').head(8)
g = sns.catplot(x = "index", y = "Medal", data = totalSilverMedals,
               height = 5, kind = "bar", palette="rocket")
g.despine(left = True)
g.set_xlabels("Top 8 Countries")
g.set_ylabels("Number of Medals")
plt.title('Silver Medals per country')
```
